```python
# Imports
import requests
```

```python
# Ollama endpoint
url = "http://localhost:11434/api/generate"
```

```python
# 3. Function to generate code using different LLMs with given prompt
def generate_code(model, prompt):
    # trying system prompt which will be a special instruction from my side to add more
    # context behaviour or constraints if i need
    system_prompt = "You are a helpful assistant that generates clean, well-documented Python code."

    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False,
        "system_prompt": system_prompt,
        "temperature": 0.2, # trying little strict than chaotic
        # "max_tokens": 1000,
    }

    if system_prompt:
        payload["system"] = system_prompt

    # request to local server
    response = requests.post(url, json=payload)

    if response.status_code == 200:
        return response.json()['response'].strip()
    else:
        raise Exception(f"Failed to generate: {response.text}")
```

```python
# A basic code generation example
# prompt = "Write a code for finding whether the given number is odd or even."

# Little complex code generation example > generating a Django boilerplate
prompt = """
Generate a basic Django project boilerplate named "generated_project" with one app named "core". Include:
- Directory structure
- `manage.py`
- `settings.py`, `urls.py`, `wsgi.py` inside the project folder
- `models.py`, `views.py`, `urls.py` inside the core app
- Add basic dummy code in each file (like one view, one model)
- Include a `requirements.txt` with Django
Format everything as a code block per file, like:

generated_project/
├── manage.py
├── generated_project/
│   ├── __init__.py
│   ├── settings.py
│   ├── urls.py
│   ├── wsgi.py
├── core/
│   ├── __init__.py
│   ├── views.py
│   ├── models.py
│   ├── urls.py
requirements.txt
"""
```
