````python
import os
import requests
from pathlib import Path
from IPython.display import display, Markdown
````

````python
OLLAMA_BASE_URL = "http://localhost:11434/api/generate"
MODEL_NAME = "deepseek-r1:14b"
````

````python
def build_prompt(file_path, file_content):
    # giving system prompt here itself
    return f"""
    You are a code documentation assistant.

    Analyze the following Python file and generate:
    1. Google-style docstrings for all functions and classes.
    2. Inline comments for complex or non-obvious logic.
    3. A one-paragraph summary of the file’s purpose.

    Respond with just the documented code and the summary.

    File: {file_path}

    ```python
    {file_content}

    ```
    """
````

````python
def generate_doc(prompt):
    response = requests.post(
        OLLAMA_BASE_URL,
        json={
            "model": MODEL_NAME,
            "prompt": prompt,
            "stream": False
        }
    )
    return response.json().get("response", "").strip()
````

````python
def document_python_files(project_root: str):
    # within the same project folder
    project_root = Path(project_root)

    # going with python files for now
    # py_files = list(project_root.rglob("*.py"))

    # trying .ipynb file
    py_files = list(project_root.rglob("*.ipynb"))

    print(f"Found {len(py_files)} Python files in {project_root}\n")

    for file_path in py_files:
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                content = f.read()

            prompt = build_prompt(file_path.name, content)
            print(f"Documenting: {file_path.name}")
            doc_output = generate_doc(prompt)

            # Display
            display(Markdown(f"### **{file_path.name}**"))
            display(Markdown("#### Original Code"))
            display(Markdown(f"```python\n{content[:1000]}\n...```" if len(content) > 1000 else f"```python\n{content}```"))

            display(Markdown("#### Documented Output"))
            display(Markdown(f"```python\n{doc_output}```"))

        except Exception as e:
            print(f"Error reading {file_path}: {e}")
````
